```python
#Implement a student library system using OOPs where students can borrow a books from the lists of books.
#Create a seprate library and class. your program must be menu driven.
#You are free to choose methods thad and attributes of your choice to implement this functionality.
```

```python
class Library:
    def __init__(self):
        self.books = [
            "Python Programming",
            "Data Science Basics",
            "Machine Learning",
            "Artificial Intelligence",
            "Database Management"
        ]

    def display_books(self):
        if len(self.books) == 0:
            print("\nNo books available in the library.")
        else:
            print("\nAvailable Books:")
            for i, book in enumerate(self.books, start=1):
                print(f"{i}. {book}")

    def borrow_book(self, book_name):
        if book_name in self.books:
            self.books.remove(book_name)
            return True
        return False

    def return_book(self, book_name):
        self.books.append(book_name)
        print(f"\n'{book_name}' has been returned successfully.")


class Student:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def borrow(self, library):
        book_name = input("Enter the book name you want to borrow: ")

        if library.borrow_book(book_name):
            self.borrowed_books.append(book_name)
            print(f"\n{self.name} borrowed '{book_name}' successfully.")
        else:
            print("\nSorry! This book is not available.")

    def return_book(self, library):
        book_name = input("Enter the book name you want to return: ")

        if book_name in self.borrowed_books:
            self.borrowed_books.remove(book_name)
            library.return_book(book_name)
        else:
            print("\nYou have not borrowed this book.")

    def view_borrowed_books(self):
        if len(self.borrowed_books) == 0:
            print("\nNo books borrowed.")
        else:
            print("\nBorrowed Books:")
            for book in self.borrowed_books:
                print(book)


# Main Program
library = Library()

student_name = input("Enter Student Name: ")
student = Student(student_name)

while True:
    print("\n===== STUDENT LIBRARY SYSTEM =====")
    print("1. Display Available Books")
    print("2. Borrow Book")
    print("3. Return Book")
    print("4. View Borrowed Books")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        library.display_books()

    elif choice == "2":
        student.borrow(library)

    elif choice == "3":
        student.return_book(library)

    elif choice == "4":
        student.view_borrowed_books()

    elif choice == "5":
        print("\nThank you for using the Library System!")
        break

    else:
        print("\nInvalid choice! Please try again.")
```
